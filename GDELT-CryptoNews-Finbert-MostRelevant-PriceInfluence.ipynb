```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_excel("GDELT-CryptoNews-BQ-and-Bitcoin-ClosingPrice-WithNews-WithFinBert-MostRelevant-ByDate.xlsx")
```

```python
df
```

```python
df["price_increase"] = ((df["finbert"] == "positive") & ((df["ClosingPrice"] > df["ClosingPrice"].shift(1)) |
                                  (df["ClosingPrice"] < df["ClosingPrice"].shift(-1))))
```

```python
df["price_increase"].value_counts()
```

```python
df["price_drop"] = ((df["finbert"] == "negative") & ((df["ClosingPrice"] < df["ClosingPrice"].shift(1)) |
                                  (df["ClosingPrice"] > df["ClosingPrice"].shift(-1))))
```

```python
df["price_drop"].value_counts()
```

```python
df["price_irrelevant"] = df["finbert"] == "neutral"
```

```python
df["price_irrelevant"].value_counts()
```

```python
df["price_unexpected"] = ((df["price_increase"] == False) & (df["price_drop"] == False) & 
                          (df["price_irrelevant"] == False))
```

```python
df["price_unexpected"].value_counts()
```

```python
df
```

```python
df.to_excel("GDELT-CryptoNews-FinBert-MostRelevant-PriceInfluence.xlsx",index=False)
```
