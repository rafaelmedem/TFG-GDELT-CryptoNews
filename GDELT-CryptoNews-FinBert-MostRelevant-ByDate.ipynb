```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_excel("GDELT-CryptoNews-BQ-and-Bitcoin-ClosingPrice-WithNews-WithFinBert.xlsx")
```

```python
df
```

```python
df["max_value"] = df[["finbert_positive","finbert_negative","finbert_neutral"]].max(axis=1)
```

```python
df
```

```python
df["finbert"] = df[["finbert_positive","finbert_negative","finbert_neutral"]].idxmax(axis=1)
```

```python
df
```

```python
df["finbert"].replace({"finbert_positive":"positive","finbert_negative":"negative",
                       "finbert_neutral":"neutral"},inplace=True)
```

```python
df
```

```python
df.to_excel("GDELT-CryptoNews-BQ-and-Bitcoin-ClosingPrice-WithNews-WithFinBert_v2.xlsx",index=False)
```

```python
new_df = df.loc[df.groupby("Date")["max_value"].idxmax(), :]
```

```python
new_df
```

```python
new_df.to_excel("GDELT-CryptoNews-BQ-and-Bitcoin-ClosingPrice-WithNews-WithFinBert-MostRelevant-ByDate.xlsx",
                index=False)
```
